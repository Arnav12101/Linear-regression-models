# LOGISTIC REGRESSION ON IRIS DATASET

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('E:/Machine learning/Linear regression/datasets/iris.csv')
```

```python
data.head()
```

```python
data
```

```python
data.describe
```

```python
l1 = data['SepalLengthCm'].tolist()
l2 = data['Species'].tolist()
```

```python
x = np.array(l1).reshape(-1,1)
y = np.array(l2)
```

```python
x
```

```python
y
```

```python
model = LogisticRegression().fit(x,y)
```

```python
y_pred = model.predict(x)
print('Prediction:- ', y_pred)
```

```python
y_pred = model.predict([[5.0]])
print(y_pred)
```

```python
y_pred = model.predict([[6.9]])
print(y_pred)
```

```python
print('Coeff:- ', model.coef_)
print('\n')
print('Intercept:- ', model.intercept_)
print('\n')
print('R^2:- ',model.score(x,y))
```
